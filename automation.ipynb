# Read and show the data 

```python
import pandas as pd
df = pd.read_csv("./daily_report_python.csv")
df_att = pd.read_csv("./data.csv")
df
```

# spliting the dataframes 

```python
df_1 , df_2 = df[:1] , df[1:]
df_2.columns = df_2.iloc[0]
df_2.drop(df_2.index[0], inplace=True)
df_2
```

```python
meeting_duration = int(df_1["Duration (Minutes)"][0])
meeting_id = df_1["Meeting ID"][0]
meeting_id 
```

```python
df_2["Total"] = df_2["Duration (Minutes)"].astype("int")
df_2["Student ID"]= df_2["User Email"].str.extract(pat='(^[0-9]*)')
df_2["Student ID"]
```

# change the type to intager 

```python
df_3=df_2
df_2 = df_2.groupby("Student ID", as_index=False).sum(numeric_only=True)
df_2
```

```python
df_2["Status"] = ['Present' if x >= meeting_duration-10 else "Absent" if x <meeting_duration - 20 else "Late" for x in df_2["Total"]]
df_2
```

```python
df_3 = df_3.groupby("Name (Original Name)", as_index=False).sum(numeric_only=True)
df_3["Status"] = ['Present' if x >= meeting_duration-10 else "Absent" if x <meeting_duration - 20 else "Late" for x in df_3["Total"]]
df_att["Name (Original Name)"] = df_3["Name (Original Name)"]
df_att[meeting_id] = df_3["Status"]

df_att.to_csv("data.csv")
df_att
```
